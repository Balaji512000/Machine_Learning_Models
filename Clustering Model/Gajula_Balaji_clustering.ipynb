### Implementing Clustering Algorithms:

# DBSCAN MODEL

```python
from sklearn.cluster import DBSCAN     # importing dbscan for building model from scikitlearn
dbscan_model=DBSCAN(eps=3,min_samples=4,metric="euclidean")   #eps- radius btwn 2 points, #euclidean-straightline distance between 2 points
dbscan_model
```

```python
dbscan_model.fit(x_scaled)  #fitted the model with scaled values of x
```

```python
labels_db = dbscan_model.labels_ 
labels_db
```

```python
df.head(10)
```

```python
df.columns
```

```python
df["db_clusters"]=labels_db
```

```python
df.columns
```

```python
y_means = dbscan_model.fit_predict(x_scaled)
y_means
```

```python
from sklearn.metrics import silhouette_score   #silhouette_score=  -1 to 1 better score indicates better clustering 

# Calculating silhouette score   
score = silhouette_score(x_scaled, labels_db)          #-1 for bad clustering, 0 for some data point clustering moved to another cluster
print(f"Silhouette Score for dbscan: {score}")                    # 1 for good clustering
```

## Silhouette Score for dbscan: 0.5128054200759178

```python
import numpy as np
y_means = dbscan_model.fit_predict(x_scaled)   #fitted model
plt.figure(figsize=(7,5))
colors = ['pink', 'yellow', 'cyan', 'magenta', 'orange', 'blue', 'red', 'black', 'violet', 'green']  # colors
unique_clusters = np.unique(y_means)  
for cluster_id in unique_clusters:
    if cluster_id == -1:
        # Noise points (optional, can be skipped or represented differently)
        cluster_color = 'gray'
        label = 'Noise'
    else:
        cluster_color = colors[cluster_id % len(colors)]
        label = f"cluster{cluster_id+1}"
    mask = y_means == cluster_id
    plt.scatter(x_scaled[mask, 0], x_scaled[mask, 1], s=50, c=cluster_color, label=label)
plt.title('Clusters of data')
plt.legend()
plt.show()
```

# K-Means Model

```python
from sklearn.cluster import KMeans  #importing kmeans from sckit learn
wcss=[]
for i in range(1,11):  #range for clusters
  kmeans_model=KMeans(n_clusters=i,init="k-means++",random_state=1)
  kmeans_model.fit(x)            #fitting the model
  wcss.append(kmeans_model.inertia_)
# kmeans inertia measures how tightly the data points are clustered around the centroids.
# elbow methods.
```

```python
kmeans_model=KMeans(n_clusters=5,init="k-means++",random_state=1)
y_kmeans=kmeans_model.fit_predict(x)
y_kmeans
```

```python
labels_kmeans = kmeans_model.labels_ 
labels_kmeans
```

```python
#Elbow method kmeans
import matplotlib.pyplot as plt
plt.plot(range(1, 11), wcss, marker='o')
plt.title("Elbow Method K-Means")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
df["kmeans_clusters"]=labels_kmeans
```

```python
df.columns
```

```python
# Silhouette Score
from sklearn.metrics import silhouette_score
print("KMeans Silhouette Score:", silhouette_score(x,y_kmeans))
```

## KMeans Silhouette Score: 0.5481756025079243

```python
import matplotlib.pyplot as plt

plt.title("KMEANS CLUSTER (5 CLUSTERS)")

# Use .loc with boolean mask, then select columns appropriately
plt.scatter(x.loc[y_kmeans == 0, x.columns[0]], x.loc[y_kmeans == 0, x.columns[1]], s=60, c='red', label='cluster1')
plt.scatter(x.loc[y_kmeans == 1, x.columns[0]], x.loc[y_kmeans == 1, x.columns[1]], s=60, c='blue', label='cluster2')
plt.scatter(x.loc[y_kmeans == 2, x.columns[0]], x.loc[y_kmeans == 2, x.columns[1]], s=60, c='green', label='cluster3')
plt.scatter(x.loc[y_kmeans == 3, x.columns[0]], x.loc[y_kmeans == 3, x.columns[1]], s=60, c='violet', label='cluster4')
plt.scatter(x.loc[y_kmeans == 4, x.columns[0]], x.loc[y_kmeans == 4, x.columns[1]], s=60, c='yellow', label='cluster5')

# For centroids: kmeans.cluster_centers_ is a numpy array, so direct indexing works
plt.scatter(kmeans_model.cluster_centers_[:, 0], kmeans_model.cluster_centers_[:, 1], s=100, c='black', label='Centroids')

plt.legend()
plt.show()
```

```python
from sklearn import preprocessing   #preprocessing imported
import scipy.cluster.hierarchy as sch    #hierarchy imported from scipy
from sklearn.cluster import AgglomerativeClustering   #imported agglomerativeclustering  from sklearn
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
```

```python
link = sch.linkage(x_scaled, method='ward')  
```

```python
plt.figure(1, figsize = (16 ,8))
dendrogram = sch.dendrogram(sch.linkage(x_scaled, method="ward"))

plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
from scipy.cluster.hierarchy import dendrogram as dendrogram_func
linkage_methods = ['single', 'complete', 'average', 'ward']

for method in linkage_methods:
    plt.figure(figsize=(10, 5))
    Z = linkage(x_scaled, method=method)
    dendrogram(Z, truncate_mode="level", p=5)
    plt.title(f"Dendrogram using {method} linkage")
    plt.xlabel("Sample Index")
    plt.ylabel("Distance")
    plt.show()
```
