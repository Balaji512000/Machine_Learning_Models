```python
y=df["Type"]
y
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
imputer_num = SimpleImputer(strategy='mean')
df = imputer_num.fit_transform(df)
```

```python
scaler = StandardScaler()
df = scaler.fit_transform(df)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df,y,test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(random_state=42,n_jobs=-1,max_depth=5,n_estimators=100,oob_score=True)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
model.oob_score_
```

```python
from sklearn.metrics import confusion_matrix,classification_report
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
print(classification_report(y_test,y_pred))
```

```python
# Accuracy - 98% - Excellent model
```

```python
sns.heatmap(cm,annot=True)
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(x_train, y_train)
```

```python
y_pred_rf = rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print("Random Forest Accuracy (Bagging):", accuracy_score(y_test, y_pred_rf))
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

# Use 'estimator' instead of 'base_estimator'
base_estimator = DecisionTreeClassifier(max_depth=1)
ada = AdaBoostClassifier(estimator=base_estimator, n_estimators=100, random_state=42)
ada.fit(x_train, y_train)
y_pred_ada = ada.predict(x_test)
```

```python
print("AdaBoost Accuracy (Boosting):", accuracy_score(y_test, y_pred_ada))
```
