#### 1)What are some common hyperparameters of decision tree models, and how do they affect the model's performance?


1. criterion: How to measure split quality (e.g., "gini" or "entropy").
2. max_depth: Maximum tree depth; controls complexity. Too deep = overfitting, too shallow = underfitting.
3. min_samples_split: Minimum samples needed to split a node; higher values make the tree simpler.
4. min_samples_leaf: Minimum samples at a leaf; larger values prevent tiny leaves and reduce overfitting.
5. max_features: Number of features to consider at each split; fewer features can add randomness and prevent overfitting.

Tuning these helps balance bias and variance to improve model accuracy and generalization.

### 2)What is the difference between the Label encoding and One-hot encoding?

Label Encoding vs One-hot Encoding (Brief Summary):

1. Label Encoding: Converts each category into a unique integer value (e.g., Red=0, Green=1, Blue=2). It encodes the categorical feature as a single column with integers. Best when categories have an ordinal relationship (order matters). It does not increase the number of features.

2. One-hot Encoding: Converts the categorical feature into multiple binary columns (one column per category). Each row has a 1 in the column for its category and 0 in others. Best for nominal data with no order. It increases the number of features because it creates a column for each category.

#### Main difference: Label Encoding produces one integer-coded feature, while One-hot Encoding creates multiple binary features representing categories without implying any order.

#### This makes Label Encoding simpler but can mislead some algorithms that interpret the integers as ordered, while One-hot Encoding avoids this but can increase feature dimensionality.
